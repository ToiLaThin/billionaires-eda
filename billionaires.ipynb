```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### EDA

```python
data = pd.read_csv('./billionaires.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
cols_keeping = [
    'position', 'wealth', 'industry', 'full_name', 'age',
    'city_of_residence', 'country_of_residence', 'gender', 'continent'
]
data = data[cols_keeping]
data.head()
```

```python
data['wealth'].hist()
```

#### Which industry appear most

```python
data['industry'].hist(bins=50)
plt.xticks(rotation=90)
plt.show()
```

#### Top 10 industries gather most wealth

```python
where = data.groupby('industry')['wealth'].sum().sort_values(ascending=False).head(10)
where.plot(kind='bar')
where.head(10)
```

#### Median ,mode age of billionaires from each industry

```python
where = data.groupby('industry')['age'].agg(['mean', 'median']).sort_values(by='mean', ascending=False).head(10)
where.plot(kind='bar')
where.head(10)
```

#### Top 5 country have most billionaires

```python
where = data.groupby('country_of_residence')['position'].agg(['count']).sort_values(by='count', ascending=False).head(10)
where.plot(kind='bar')
where.head(10)
```

#### For india, top 3 industry have most millionaires

```python
where = data[data['country_of_residence'] == 'India'].groupby('industry')['position'].agg(['count']).sort_values(by='count', ascending=False).head(3)
where.plot(kind='bar')
where.head(3)
```

#### For india, top 3 industry have richest billionaires

```python
where = data[data['country_of_residence'] == 'India'].groupby('industry')['wealth'].agg(['max']).sort_values(by='max', ascending=False).head(3)
where.plot(kind='bar')
where.head(3)
```
